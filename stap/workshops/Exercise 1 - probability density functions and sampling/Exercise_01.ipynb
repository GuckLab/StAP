# Question 2:

```python
n=100
res = []
for i in range(1,n+1):
    res.append(np.sum(np.random.randint(1,high=7,size = 2)))
```

```python
res.count(7)*500-100*(n-res.count(7))
```

```python
def game1(n):
    res = [];
    for j in range(0,n):
            res.append(np.sum(np.random.randint(1,high=7,size = 2)))
    return res.count(7)*500-100*(n-res.count(7))
```

```python
game1(0)
```

```python
m= 100;
res2 = np.zeros((100,m+1))
for j in range(0,100):
    for i in range(1,m):
        res2[j,i]= game1(j)
```

```python
means = []
sds = []
for j in range (1,m-1):
    means.append(np.mean(res2[j,:]))
    sds.append(np.std(res2[j,:]))
```

```python
nbm =  np.linspace(1,100,100)
theosd = 1/6*np.sqrt(5)*np.sqrt(500**2*nbm + 100**2*nbm)
theom = 0*nbm
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(means)
ax.plot(sds)
ax.plot(nbm,theosd,'r')
ax.plot(nbm,theom,'r')
```

```python
def game2(val):
    return np.prod(np.random.randint(1,high=7,size = 2))-val
```

```python
game2(10)
```

```python
m= 100;
res2 = np.zeros((100,m+1))
for j in range(0,100):
    for i in range(1,m):
        res2[j,i]= game2(10)
```

```python
means = []
sds = []
for j in range (1,m-1):
    means.append(np.mean(res2[j,:]))
    sds.append(np.std(res2[j,:]))
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(means)
ax.plot(sds)
```
